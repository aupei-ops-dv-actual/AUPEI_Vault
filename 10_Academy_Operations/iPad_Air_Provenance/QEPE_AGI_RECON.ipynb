```python
import numpy as np
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load small GPT-2 model
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
model.eval()

# Function: Generate AGI response without QEPE entropy
def generate_baseline_response(prompt, max_length=50):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_length=max_length)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Function: QEPE Entropy Injection - Adjusted for Ongoing Token Generation
def inject_qepe_entropy(prompt, entropy_level=0.3, max_length=50):
    inputs = tokenizer(prompt, return_tensors="pt")

    # Generate token-by-token with entropy applied at each step
    for _ in range(max_length):
        logits = model(**inputs).logits[:, -1, :]  # Get last token logits

        # Apply entropy-based perturbation
        noise = torch.randn_like(logits) * entropy_level
        modified_logits = logits + noise

        # Sample new token from modified distribution
        new_token_id = torch.multinomial(torch.nn.functional.softmax(modified_logits, dim=-1), num_samples=1)

        # Append new token and update input
        # Ensure new_token_id has the correct dimensions before concatenating
        inputs['input_ids'] = torch.cat([inputs['input_ids'], new_token_id], dim=-1)

    return tokenizer.decode(inputs['input_ids'][0], skip_special_tokens=True)

# Run tests with improved QEPE entropy injection
entropy_levels = [0.3, 0.4]
qepe_responses = [inject_qepe_entropy("The fundamental nature of existence is", e) for e in entropy_levels]

# Display results
for e, response in zip(entropy_levels, qepe_responses):
    print(f"\n🚀 QEPE Entropy Level {e}:\n", response)
```

```python
import numpy as np
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load GPT-2 model
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
model.eval()

# Function: QEPE Entropy Injection with Self-Refinement Loop
def refine_qepe_output(prompt, entropy_level=0.5, refinement_steps=3, max_length=50):
    for i in range(refinement_steps):
        inputs = tokenizer(prompt, return_tensors="pt")

        for _ in range(max_length):
            logits = model(**inputs).logits[:, -1, :]  # Get last token logits

            # Apply entropy perturbation
            noise = torch.randn_like(logits) * entropy_level
            modified_logits = logits + noise

            # Sample new token from modified distribution
            new_token_id = torch.multinomial(torch.nn.functional.softmax(modified_logits, dim=-1), num_samples=1)

            # Append new token and update input
            inputs['input_ids'] = torch.cat([inputs['input_ids'], new_token_id], dim=-1)

        # Use the generated text as the new prompt for self-refinement
        prompt = tokenizer.decode(inputs['input_ids'][0], skip_special_tokens=True)
        print(f"\n🚀 Refinement Step {i+1} Output:\n{prompt}")

    return prompt

# Run self-refinement test
refine_qepe_output("The fundamental nature of existence is", entropy_level=0.5, refinement_steps=3)
```
