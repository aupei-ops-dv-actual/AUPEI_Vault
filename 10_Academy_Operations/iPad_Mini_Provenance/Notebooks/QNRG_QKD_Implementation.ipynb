Main 7dU QNRG system




```python
import numpy as np
import hashlib

class QuantumRandomNumberGenerator:
    def __init__(self, seed=None):
        """Initialize QNRG with an optional seed for testing."""
        self.seed = seed or np.random.rand()

    def generate(self, size=256):
        """Generate a quantum-derived random bit sequence."""
        np.random.seed(int(self.seed * 1e6))
        bits = np.random.randint(0, 2, size)
        return bits

    def hash_output(self, bits):
        """Apply SHA-3 hashing to enhance randomness."""
        bit_string = ''.join(map(str, bits))
        return hashlib.sha3_256(bit_string.encode()).hexdigest()

# Instantiate and test
qrng = QuantumRandomNumberGenerator()
random_bits = qrng.generate()
hashed_output = qrng.hash_output(random_bits)

print("Quantum Random Bits:", random_bits[:20])  # Show first 20 bits
print("Hashed Output:", hashed_output)
```

KQD Integration




```python
class QKD_Protocol:
    def __init__(self, qrng):
        """Integrates QNRG with Quantum Key Distribution."""
        self.qrng = qrng

    def generate_key(self, length=128):
        """Generate a secure quantum key for encryption."""
        key_bits = self.qrng.generate(length)
        return ''.join(map(str, key_bits))

    def encrypt_message(self, message, key):
        """XOR encrypt a message with a quantum-generated key."""
        message_bits = ''.join(format(ord(c), '08b') for c in message)
        encrypted_bits = ''.join(str(int(m) ^ int(k)) for m, k in zip(message_bits, key))
        return encrypted_bits

# Test QKD with QNRG
qkd = QKD_Protocol(qrng)
secure_key = qkd.generate_key()
encrypted_message = qkd.encrypt_message("Quantum Security!", secure_key)

print("Secure Key:", secure_key[:20])  # First 20 bits
print("Encrypted Message:", encrypted_message[:50])  # Show snippet
```

Honey Bells Security Mechanism




```python
class HoneyBellSecurity:
    def __init__(self, sensitivity=0.1):
        """Initialize Honey Bell Security trapdoor mechanism."""
        self.sensitivity = sensitivity

    def detect_intrusion(self, key1, key2):
        """Compare two quantum-generated keys to detect tampering."""
        mismatches = sum(1 for b1, b2 in zip(key1, key2) if b1 != b2)
        error_rate = mismatches / len(key1)
        return error_rate > self.sensitivity  # Trigger alert if threshold exceeded

# Test intrusion detection
hbs = HoneyBellSecurity()
intrusion_detected = hbs.detect_intrusion(qkd.generate_key(), qkd.generate_key())

print("Intrusion Detected:", intrusion_detected)
```

Entropy Analysis & Trapdoor Defense




```python
import scipy.stats as stats

class EntropyAnalyzer:
    def __init__(self, data):
        """Analyze entropy of a given bit sequence."""
        self.data = data

    def compute_entropy(self):
        """Calculate Shannon entropy."""
        probs = np.bincount(self.data) / len(self.data)
        return -np.sum(p * np.log2(p) for p in probs if p > 0)

# Evaluate entropy of QNRG output
analyzer = EntropyAnalyzer(qrng.generate())
entropy_value = analyzer.compute_entropy()

print("QNRG Entropy:", entropy_value)
```
