```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Define the neutrino-entanglement velocity function
def neutrino_entanglement_velocity(r, alpha, beta, gamma):
    return np.sqrt(1 / r + alpha * np.exp(-beta * r) * np.cos(gamma * r))

# Optimized parameters
alpha_opt = 1.0351e4
beta_opt = -0.0802
gamma_opt = 0.0400

# Directory containing SPARC data files
data_dir = 'path_to_extracted_sparc_data'

# Iterate through data files
for filename in os.listdir(data_dir):
    if filename.endswith('.dat'):
        # Load data
        data = np.loadtxt(os.path.join(data_dir, filename), comments='#')
        r_obs = data[:, 0]  # Radial distances (kpc)
        v_obs = data[:, 1]  # Observed velocities (km/s)

        # Compute model velocities
        v_model = neutrino_entanglement_velocity(r_obs, alpha_opt, beta_opt, gamma_opt)

        # Plot observed vs. model rotation curves
        plt.figure(figsize=(8, 6))
        plt.plot(r_obs, v_obs, 'bo-', label='Observed Rotation Curve')
        plt.plot(r_obs, v_model, 'r--', label='Neutrino-Entanglement Model')
        plt.xlabel('Radial Distance (kpc)')
        plt.ylabel('Rotational Velocity (km/s)')
        plt.title(f'Rotation Curve Comparison for {filename[:-4]}')
        plt.legend()
        plt.grid(True, linestyle="--")
        plt.show()
```
