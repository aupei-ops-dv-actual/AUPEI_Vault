## **🔹 Loading Core QNRG & QKD Components**

This section loads the main components of the **7dU QNRG & QKD system**, including:
- **QNRG**: Quantum-Inspired Random Number Generator
- **QKD Security Features**: Honey Bell Entanglement, Decoy States, and Trapdoor Mechanisms
- **Entropy Enhancement Modules**: Nonlinear transforms, quantum diffusion, and entropy reconciliation

```python
# 🚀 Loading Core QNRG & QKD Components

import numpy as np
from hashlib import sha3_256
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

# Quantum-Inspired Random Number Generator (QNRG)
def generate_qnrg_entropy(length=256):
    """
    Generates entropy using a quantum-inspired approach with Qiskit.
    """
    # Initialize quantum circuit with the number of qubits equal to 'length'
    qc = QuantumCircuit(length, length)

    # Apply Hadamard gate to each qubit to create superposition
    for qubit in range(length):
        qc.h(qubit)

    # Measure each qubit
    qc.measure(range(length), range(length))

    # Use Aer's simulator
    simulator = AerSimulator()
    compiled_circuit = transpile(qc, simulator)
    result = simulator.run(compiled_circuit).result()
    counts = result.get_counts()

    # Convert the result to a binary string
    binary_string = max(counts, key=counts.get)

    # Convert binary string to an array of integers
    entropy_values = np.array([int(bit) for bit in binary_string], dtype=np.uint8)

    return entropy_values

# Honey Bell Entanglement Simulation for QKD
def honey_bell_qkd(length=128):
    """
    Simulates Honey Bell entangled quantum key distribution.
    """
    # Initialize quantum circuit with 2 qubits and 2 classical bits
    qc = QuantumCircuit(2, 2)

    # Create Bell state
    qc.h(0)
    qc.cx(0, 1)

    # Measure qubits
    qc.measure([0, 1], [0, 1])

    # Use Aer's simulator
    simulator = AerSimulator()

    alice_key = []
    bob_key = []

    for _ in range(length):
        compiled_circuit = transpile(qc, simulator)
        result = simulator.run(compiled_circuit).result()
        counts = result.get_counts()
        measured_bits = max(counts, key=counts.get)
        alice_bit, bob_bit = int(measured_bits[0]), int(measured_bits[1])
        alice_key.append(alice_bit)
        bob_key.append(bob_bit)

    return np.array(alice_key, dtype=np.uint8), np.array(bob_key, dtype=np.uint8)

# Quantum Diffusion Mechanism for Entropy Smoothing
def quantum_diffuse(data):
    """
    Applies a quantum-inspired diffusion transform to improve entropy spread.
    """
    transformed_data = np.fft.ifft(np.fft.fft(data) * np.exp(1j * np.random.rand(len(data)))).real
    return np.abs(transformed_data * 255).astype(np.uint8)

# Trapdoor Detection Mechanism
def trapdoor_security_check(qkd_key):
    """
    Simulates a trapdoor security mechanism detecting eavesdropping patterns.
    """
    return np.var(qkd_key)  # Higher variance may indicate interference

print("✅ Core QNRG & QKD components loaded successfully.")
```

	•	Honey Bell Entanglement Simulation (honey_bell_qkd): This function simulates the creation of entangled qubit pairs (Bell states) and measures them to generate correlated keys for Alice and Bob.
	•	Quantum Diffusion Mechanism (quantum_diffuse): This function applies a Fourier transform-based diffusion to the data to enhance entropy distribution.

```python
import numpy as np
import scipy.stats as stats

# Generate a random bitstream for testing (replace with actual QNRG output)
bitstream = np.random.randint(0, 2, 100000)

# **1️⃣ Monobit (Frequency) Test**
def monobit_test(bits):
    """Checks if the number of 1s and 0s are balanced."""
    n = len(bits)
    s = sum(2*x - 1 for x in bits)  # Convert 0s to -1 and sum
    test_stat = abs(s) / np.sqrt(n)
    p_value = stats.norm.sf(test_stat) * 2  # Two-tailed test
    return {"Test": "Monobit", "Statistic": test_stat, "p-value": p_value}

# **2️⃣ Runs Test**
def runs_test(bits):
    """Counts the number of runs of 1s and 0s and compares to expected."""
    n = len(bits)
    pi = sum(bits) / n  # Proportion of 1s
    tau = 2 / np.sqrt(n)  # Threshold for randomness
    if abs(pi - 0.5) >= tau:
        return {"Test": "Runs", "Error": "Fails balance condition"}

    runs = 1 + sum(bits[i] != bits[i+1] for i in range(n - 1))
    expected_runs = 2 * n * pi * (1 - pi)
    variance = expected_runs * (expected_runs - 1) / (n - 1)
    test_stat = abs(runs - expected_runs) / np.sqrt(variance)
    p_value = stats.norm.sf(test_stat) * 2  # Two-tailed test
    return {"Test": "Runs", "Statistic": test_stat, "p-value": p_value}

# **3️⃣ Longest Run of Ones Test**
def longest_run_test(bits):
    """Checks if the longest sequence of 1s is within expected limits."""
    n = len(bits)
    max_run = 0
    current_run = 0
    for bit in bits:
        if bit == 1:
            current_run += 1
            max_run = max(max_run, current_run)
        else:
            current_run = 0

    expected_max_run = np.log2(n)
    test_stat = abs(max_run - expected_max_run) / np.sqrt(expected_max_run)
    p_value = stats.norm.sf(test_stat) * 2  # Two-tailed test
    return {"Test": "Longest Run", "Statistic": test_stat, "p-value": p_value}

# **4️⃣ Block Frequency Test**
def block_frequency_test(bits, block_size=128):
    """Divides the bitstream into blocks and checks uniformity of 1s."""
    n = len(bits)
    num_blocks = n // block_size
    block_sums = [sum(bits[i * block_size:(i+1) * block_size]) for i in range(num_blocks)]
    pi_vals = np.array(block_sums) / block_size  # Proportion of 1s per block
    test_stat = np.sum((pi_vals - 0.5) ** 2) * 4 * block_size
    p_value = stats.chi2.sf(test_stat, df=num_blocks-1)
    return {"Test": "Block Frequency", "Statistic": test_stat, "p-value": p_value}

# **Run All Tests**
test_results = [
    monobit_test(bitstream),
    runs_test(bitstream),
    longest_run_test(bitstream),
    block_frequency_test(bitstream)
]

# **Display Results**
import pandas as pd
df_results = pd.DataFrame(test_results)
print("\n🚀 NIST-Based Randomness Test Results 🚀\n")
print(df_results.to_string(index=False))
```

```python
import numpy as np
import scipy.stats as stats
import time

# 🚀 **Bit Bias Test**
def bit_bias_test(bitstream):
    ones_count = np.sum(bitstream)
    total_bits = len(bitstream)
    bit_bias_stat = ones_count / total_bits
    p_value = stats.binomtest(ones_count, total_bits, 0.5, alternative='two-sided').pvalue
    return {"Bit Bias": bit_bias_stat, "p-value": p_value}

# 🚀 **Runs Test**
def runs_test(bitstream):
    n = len(bitstream)
    runs = 1 + np.sum(bitstream[:-1] != bitstream[1:])
    expected_runs = (2 * n - 1) / 3
    variance_runs = (16 * n - 29) / 90
    z_score = (runs - expected_runs) / np.sqrt(variance_runs)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    return {"Runs Test Z-Score": z_score, "p-value": p_value}

# 🚀 **Autocorrelation Test**
def autocorrelation_test(bitstream, lag=1):
    n = len(bitstream)
    mean = np.mean(bitstream)
    autocorr = np.correlate(bitstream - mean, bitstream - mean, mode='full')[n-1:] / (np.var(bitstream) * n)
    return {"Autocorrelation Coefficient": autocorr[lag]}

# ✅ **Generate Entropy-Corrected Bitstream**
print("🔄 Generating entropy-corrected QNRG bitstream...\n")
start_time = time.time()
entropy_corrected_bits = np.random.randint(0, 2, 100000)  # Replace with actual QNRG output
generation_time = time.time() - start_time

# ✅ **Run Tests**
print("\n🚀 Running Randomness Tests...\n")
bit_bias_results = bit_bias_test(entropy_corrected_bits)
runs_test_results = runs_test(entropy_corrected_bits)
autocorr_results = autocorrelation_test(entropy_corrected_bits)

# ✅ **Display Results**
print("\n🚀 **Randomness Test Results** 🚀\n")
print(f"Bit Bias: {bit_bias_results}")
print(f"Runs Test: {runs_test_results}")
print(f"Autocorrelation: {autocorr_results}")
print(f"Bitstream Generation Time: {generation_time:.6f} sec")
```
