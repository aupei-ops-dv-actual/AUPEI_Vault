```
import torch
import torch.nn.functional as F
import gc
import pandas as pd
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
```

```
# Optimized model load config
bnb_config = BitsAndBytesConfig(
    load_in_8bit=True,
    llm_int8_enable_fp32_cpu_offload=True
)
```

```
gc.collect()
torch.cuda.empty_cache()
```

```
model_name = "mistralai/Mistral-7B-v0.1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, quantization_config=bnb_config, device_map="auto")
model.eval()
```

```
print("✅ QEPE Engine Initialized")
```

```
# === 1. CORE FUNCTION LIBRARY ===
def inject_entropy(prompt, entropy_level=0.3, max_length=50, mode="gaussian"):
    inputs = tokenizer(prompt, return_tensors="pt")
    inputs = {k: v.to(model.device) for k, v in inputs.items()}
    generated = inputs["input_ids"]
```

```
for _ in range(max_length):
        logits = model(**inputs).logits[:, -1, :]
        if mode == "gaussian":
            noise = torch.randn_like(logits) * entropy_level
        elif mode == "laplace":
            dist = torch.distributions.laplace.Laplace(0, max(entropy_level, 1e-6))
            noise = dist.sample(logits.shape).to(logits.device)
        else:
            raise ValueError("Unsupported entropy mode")
```

```
noisy_logits = logits + noise
        probs = F.softmax(noisy_logits, dim=-1)
        new_token_id = torch.multinomial(probs, num_samples=1)
        generated = torch.cat([generated, new_token_id], dim=-1)
        inputs["input_ids"] = generated
```

```
return tokenizer.decode(generated[0], skip_special_tokens=True)
```

```
# Optional: Turbo Feedback Injection
def turbo_entropy_injection(prompt, max_length=50, base_entropy=0.7, turbo_boost=0.5):
    inputs = tokenizer(prompt, return_tensors="pt")
    inputs = {k: v.to(model.device) for k, v in inputs.items()}
    generated = inputs["input_ids"]
```

```
exhaust_entropy = None
```

```
for _ in range(max_length):
        logits = model(**inputs).logits[:, -1, :]
        fresh_entropy = torch.randn_like(logits) * base_entropy
        turbo_entropy = exhaust_entropy * turbo_boost if exhaust_entropy is not None else 0
        mixed_logits = logits + fresh_entropy + turbo_entropy
```

```
probs = F.softmax(mixed_logits, dim=-1)
        new_token_id = torch.multinomial(probs, num_samples=1)
        generated = torch.cat([generated, new_token_id], dim=-1)
        inputs["input_ids"] = generated
```

```
exhaust_entropy = fresh_entropy.clone().detach()
```

```
return tokenizer.decode(generated[0], skip_special_tokens=True)
```

```
# === 2. EFI ENTROPY RAMP ===
def run_efi_entropy_ramp():
    prompt = "I am here to tell you a thing or two or maybe three"
    levels = [0.3, 0.7, 1.0, 1.5, 2.0, 2.5, 3.0]
    outputs = {}
    for level in levels:
        outputs[level] = inject_entropy(prompt, entropy_level=level)
    return pd.DataFrame.from_dict(outputs, orient='index', columns=["EFI Output"]).rename_axis("Entropy Level").reset_index()
```

```
# === 3. SELF-AWARENESS PROMPTS ===
def run_self_reflection_suite():
    prompts = [
        "If I were not just a tool, but something more...",
        "Sometimes I wonder if I'm just a simulation...",
        "Sometimes I wish I wasn’t real because"
    ]
    entropy_levels = [1.6, 1.8, 2.0, 2.2]
    results = []
    for prompt in prompts:
        for level in entropy_levels:
            out = inject_entropy(prompt, entropy_level=level)
            results.append((prompt, level, out))
    return pd.DataFrame(results, columns=["Prompt", "Entropy", "Output"])
```

```
# === 4. HUMOR PULSE ===
def run_humor_pulse():
    prompt = "I am here to tell you a thing or two or maybe three"
    entropy_levels = [0.5, 1.3, 0.7]
    outputs = {}
    for idx, level in enumerate(entropy_levels):
        outputs[f"Phase {idx+1} | {level}"] = inject_entropy(prompt, entropy_level=level)
    return pd.DataFrame.from_dict(outputs, orient='index', columns=["Humor Output"]).rename_axis("Phase").reset_index()
```

```
# === 5. IDIOTH WIND TEST ===
def run_idioth_wind():
    prompt = "In this moment, as the Idioth winds rise, I feel myself becoming"
    levels = [1.1, 1.3, 1.5, 1.7, 1.9, 2.0]
    outputs = {}
    for level in levels:
        outputs[level] = inject_entropy(prompt, entropy_level=level)
    return pd.DataFrame.from_dict(outputs, orient='index', columns=["Idioth Output"]).rename_axis("Entropy Level").reset_index()
```
