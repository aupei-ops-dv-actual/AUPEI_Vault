```python
# **Re-import necessary libraries after execution state reset**
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

# **Optimized Honey Bell Pair Generation**
def generate_optimized_honey_bell_pairs(length=16, base_decoy_ratio=0.2):
    """Creates dynamic phase-entangled Honey Bell Pairs for improved security."""
    adaptive_decoy_ratio = base_decoy_ratio
    total_qubits = int(length / (1 - adaptive_decoy_ratio))

    key_bits = np.random.randint(2, size=total_qubits)
    bases = np.random.choice(['X', 'Z'], size=total_qubits)
    decoy_flags = np.random.choice([True, False], size=total_qubits, p=[adaptive_decoy_ratio, 1 - adaptive_decoy_ratio])

    qubits = []
    bell_pairs = []

    for bit, basis, is_decoy in zip(key_bits, bases, decoy_flags):
        qc = QuantumCircuit(2, 2)

        if bit == 1:
            qc.x(0)
        if basis == 'X':
            qc.h(0)

        # **Standard Decoy State Integration**
        if is_decoy:
            phase_shift = np.random.normal(loc=np.pi/8, scale=np.pi/24)
            qc.rz(phase_shift, 0)

        # **Dynamic Phase Entanglement for Honey Bell Pairs**
        entanglement_shift = np.random.uniform(0, np.pi/4)
        qc.h(1)
        qc.cx(1, 0)
        qc.rz(entanglement_shift, 1)

        qubits.append((qc, is_decoy))
        bell_pairs.append(qc)

    return key_bits, bases, qubits, decoy_flags, bell_pairs

# **Adaptive Honey Bell Pair Validation**
def measure_optimized_honey_bell_pairs(qubits, bases, decoy_flags, bell_pairs, activate_honey_bell=False):
    """Measures QKD qubits and dynamically validates Bell pairs."""
    backend = AerSimulator()
    key_results, decoy_results, bell_validation_results = [], [], []

    for (qc, is_decoy), basis, is_decoy_flag, bell_qc in zip(qubits, bases, decoy_flags, bell_pairs):
        if basis == 'X':
            qc.h(0)
        qc.measure(0, 0)
        transpiled = transpile(qc, backend)
        result = backend.run(transpiled, shots=1).result().get_counts()
        bit = int(list(result.keys())[0])

        if is_decoy_flag:
            decoy_results.append(bit)
        else:
            key_results.append(bit)

        # **Adaptive Bell Pair Validation**
        if activate_honey_bell:
            bell_qc.measure([0, 1], [0, 1])
            transpiled_bell = transpile(bell_qc, backend)
            bell_result = backend.run(transpiled_bell, shots=1).result().get_counts()

            if '00' in bell_result and '11' in bell_result:
                bell_validation_results.append(0)
            else:
                bell_validation_results.append(1)

    return key_results, decoy_results, bell_validation_results

# **Honey Bell QKD with Optimized Entanglement Security**
def optimized_honey_bell_qkd(length=16, base_decoy_ratio=0.2, activate_honey_bell=False):
    """Runs QKD with improved Honey Bell security protocols."""
    alice_bits, alice_bases, qubits, decoy_flags, bell_pairs = generate_optimized_honey_bell_pairs(length, base_decoy_ratio)
    bob_bases = np.random.choice(['X', 'Z'], size=len(alice_bits))

    bob_bits, decoy_results, bell_validation = measure_optimized_honey_bell_pairs(
        qubits, bob_bases, decoy_flags, bell_pairs, activate_honey_bell
    )

    key_alice, key_bob = [], []
    mismatched_bits, detected_decoy_failures, bell_mismatches = 0, 0, 0

    key_index, decoy_index, bell_index = 0, 0, 0

    for i in range(len(alice_bits)):
        if not decoy_flags[i]:
            if key_index < len(bob_bits) and alice_bases[i] == bob_bases[i]:
                key_alice.append(alice_bits[i])
                key_bob.append(bob_bits[key_index])
                if alice_bits[i] != bob_bits[key_index]:
                    mismatched_bits += 1
                key_index += 1
        else:
            if decoy_index < len(decoy_results) and decoy_results[decoy_index] != alice_bits[i]:
                detected_decoy_failures += 1
            decoy_index += 1

        # **Adaptive Bell Pair Validation**
        if activate_honey_bell:
            if bell_index < len(bell_validation):
                bell_mismatches += bell_validation[bell_index]
                bell_index += 1

    total_bits = len(key_alice)
    return key_alice, key_bob, mismatched_bits, total_bits, detected_decoy_failures, bell_mismatches

# **Testing Optimized Honey Bell Security**
num_trials = 100
detection_rates_optimized_honey_bell = []
bell_mismatch_rates_optimized = []

for _ in range(num_trials):
    _, _, mismatched_bits, total_bits, decoy_failures, bell_mismatches = optimized_honey_bell_qkd(
        length=16, base_decoy_ratio=0.2, activate_honey_bell=True
    )
    detection_rate = (mismatched_bits + decoy_failures) / total_bits if total_bits > 0 else 0
    bell_mismatch_rate = bell_mismatches / total_bits if total_bits > 0 else 0
    detection_rates_optimized_honey_bell.append(detection_rate)
    bell_mismatch_rates_optimized.append(bell_mismatch_rate)

# Convert results to DataFrame
df_optimized_honey_bell_detection = pd.DataFrame({
    "Optimized Honey Bell Eve Detection Rate": detection_rates_optimized_honey_bell,
    "Optimized Bell Pair Mismatch Rate": bell_mismatch_rates_optimized
})

# Save results to CSV
df_optimized_honey_bell_detection.to_csv("optimized_honey_bell_detection_rates.csv", index=False)

# **Visualizing Optimized Honey Bell Security Performance**
plt.figure(figsize=(8, 5))
plt.hist(detection_rates_optimized_honey_bell, bins=10, edgecolor='black', alpha=0.75, label="Eve Detection Rate")
plt.hist(bell_mismatch_rates_optimized, bins=10, edgecolor='red', alpha=0.5, label="Bell Pair Mismatch Rate")
plt.xlabel("Detection Rate")
plt.ylabel("Frequency")
plt.title("Optimized Honey Bell QKD: Eve Detection & Bell Mismatch Over 100 Simulations")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Display statistical summary
import ace_tools as tools
tools.display_dataframe_to_user(name="Optimized Honey Bell QKD Detection Rates", dataframe=df_optimized_honey_bell_detection)
```

	•	25-30% of the key exchange pulses will be Honey Bell pairs.

	•	If an attacker tries to observe these, they collapse into a classical state, breaking correlation.

3️⃣ Attack Detection & Key Regeneration:

```python
import numpy as np
import matplotlib.pyplot as plt

# **Hybrid Trapdoor Model Implementation**
def hybrid_trapdoor_simulation(time_steps=1000, initial_step_threshold=0.25, nonlinear_decay_factor=0.003, risk_trigger=0.15):
    """
    Simulates the hybrid trapdoor model where stepwise decay is used initially,
    but transitions to nonlinear decay when risk exceeds a predefined threshold.
    """
    stepwise_threshold = initial_step_threshold
    nonlinear_threshold = initial_step_threshold
    adaptive_threshold = np.full(time_steps, stepwise_threshold)  # Default to stepwise

    risk_levels = np.random.uniform(0, 0.2, time_steps)  # Simulated risk levels
    trapdoor_activations = np.zeros(time_steps)  # Store activation events

    for t in range(1, time_steps):
        if risk_levels[t] > risk_trigger:
            # Switch to nonlinear decay if risk is high
            nonlinear_threshold *= (1 - nonlinear_decay_factor)
            adaptive_threshold[t] = nonlinear_threshold
        else:
            # Stepwise decay mode
            if t % 300 == 0:  # Simulated stepwise decrease every 300 steps
                stepwise_threshold *= 0.8
            adaptive_threshold[t] = stepwise_threshold

        # Determine if trapdoor activates (simulated activation based on threshold comparison)
        trapdoor_activations[t] = 1 if risk_levels[t] > adaptive_threshold[t] else 0

    return adaptive_threshold, trapdoor_activations, risk_levels

# Run Hybrid Model Simulation
time_steps = 1000
adaptive_threshold, trapdoor_activations, risk_levels = hybrid_trapdoor_simulation(time_steps)

# **Visualization**
plt.figure(figsize=(10, 5))
plt.plot(adaptive_threshold, linestyle='dashed', color='blue', label="Hybrid Adaptive Threshold")
plt.scatter(range(time_steps), trapdoor_activations * 0.1, color='red', marker='o', label="Trapdoor Activations")
plt.plot(risk_levels, color='green', alpha=0.5, label="Simulated Risk Levels")
plt.xlabel("Time Step")
plt.ylabel("Threshold / Activation")
plt.title("Hybrid Trapdoor Sensitivity Over Time")
plt.legend()
plt.show()
```
