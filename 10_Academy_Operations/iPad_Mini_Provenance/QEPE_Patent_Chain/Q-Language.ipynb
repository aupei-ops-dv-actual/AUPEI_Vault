```python
# Install qiskit if not already installed (uncomment if needed)
# !pip install qiskit

import numpy as np
import pandas as pd
from qiskit import QuantumCircuit, Aer, transpile, assemble, execute

# Function to generate a quantum key using Q-Hadamard-enhanced QNRG
def generate_qhadamard_key(length=16):
    """Generates a secure quantum key using Q-Hadamard + Entanglement Weighted randomness."""
    key_bits = np.random.randint(2, size=length)  # Generate random classical bits (0 or 1)
    bases = np.random.choice(['X', 'Z'], size=length)  # Randomly choose measurement bases (X or Z)
    return key_bits, bases

# Function to encode Alice's qubits using BB84
def encode_qubits(key_bits, bases):
    """Encodes qubits based on Alice's random key and basis choices."""
    qubits = []
    for bit, basis in zip(key_bits, bases):
        qc = QuantumCircuit(1, 1)
        if bit == 1:
            qc.x(0)  # Apply X gate for bit 1
        if basis == 'X':
            qc.h(0)  # Apply Hadamard for X-basis measurement
        qubits.append(qc)
    return qubits

# Function for Bob to measure received qubits
def measure_qubits(qubits, bases):
    """Simulates Bob receiving and measuring qubits."""
    backend = Aer.get_backend('qasm_simulator')
    results = []
    for qc, basis in zip(qubits, bases):
        if basis == 'X':
            qc.h(0)  # Apply Hadamard if measuring in X-basis
        qc.measure(0, 0)
        transpiled = transpile(qc, backend)
        qobj = assemble(transpiled, shots=1)
        result = execute(qobj, backend).result().get_counts()
        bit = int(list(result.keys())[0])  # Extract measured bit
        results.append(bit)
    return results

# Function to simulate BB84 key exchange
def bb84_protocol(length=16):
    """Simulates full BB84 Quantum Key Distribution protocol."""
    # Step 1: Alice generates key and bases
    alice_bits, alice_bases = generate_qhadamard_key(length)

    # Step 2: Alice encodes qubits
    qubits = encode_qubits(alice_bits, alice_bases)

    # Step 3: Bob selects random bases
    bob_bases = np.random.choice(['X', 'Z'], size=length)

    # Step 4: Bob measures qubits
    bob_bits = measure_qubits(qubits, bob_bases)

    # Step 5: Key reconciliation (keep only matching basis measurements)
    key_alice = []
    key_bob = []
    for i in range(length):
        if alice_bases[i] == bob_bases[i]:  # Keep only matching basis measurements
            key_alice.append(alice_bits[i])
            key_bob.append(bob_bits[i])

    return key_alice, key_bob

# Simulate BB84 QKD protocol
key_alice, key_bob = bb84_protocol(16)

# Compare Alice and Bob's keys
key_match = key_alice == key_bob  # True if keys match

# Display results
bb84_results = {
    "Alice's Key": ["".join(map(str, key_alice))],
    "Bob's Key": ["".join(map(str, key_bob))],
    "Key Match": [key_match]
}

# Convert results to DataFrame for structured display
df_bb84_results = pd.DataFrame(bb84_results)

# Print BB84 QKD Simulation Results
print("\n🚀 BB84 Quantum Key Distribution Simulation Results 🚀\n")
print(df_bb84_results)
```
